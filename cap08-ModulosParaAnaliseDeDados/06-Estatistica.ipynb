```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([15, 23, 63, 94, 102])
```

```python
np.mean(a) #média
```

```python
np.std(a) #desvio padrão
```

```python
np.var(a) #variancia
```

```python
plt.plot(a[:,0], a[:,1], 'o', markersize=5, alpha=0.50)
plt.plot(m[0], m[1], 'ro', markersize=10)
plt.show()
```
