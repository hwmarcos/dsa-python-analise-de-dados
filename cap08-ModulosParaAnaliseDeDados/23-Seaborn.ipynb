```python
#visualição estatistica de dados
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sea
```

```python
dados = sea.load_dataset('tips')
```

```python
dados.head()
```

```python
sea.jointplot('total_bill', 'tip', dados, kind='reg')
```

```python
sea.lmplot('total_bill', 'tip', dados, col='smoker')
```
