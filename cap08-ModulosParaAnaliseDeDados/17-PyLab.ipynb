```python
from pylab import *
```

```python
x = linspace(0, 5 , 10)
```

```python
print(x)
```

```python
y = x ** 2
```

```python
print(y)
```

```python
fig = plt.figure()
axes = fig.add_axes([1, 1, 0.8, 0.8])
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de Linha')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de Linha')

axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes2.plot(x, y, 'g')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('Gráfico de Linha')
```
